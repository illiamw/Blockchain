# 0. Entendendo os dados 

### Incluindo bibliotecas necessarias

```python
import pandas
import numpy
import scipy
from apyori import apriori  
import matplotlib.pyplot as plt  
```

### Recolhendo o banco de dados "Dotto"

##### Como pré processamento inicial foi removido o delimitador de linha "," e considerado quebra de linha como delimitador de linhas e espaço como de celulas

```python
data = pandas.read_csv("dotto.csv", header=None,sep=' ')
```

```python
data.head()
```

# 1. Pré processamento 

##### pré processamento em converter os dados da estrutura de dados panda para lista de listas na estrura python pure

```python
ListaTransacoes = []  
for i in range(0, 1716):
    print(i)
    ListaTransacoes.append([str(data.values[i,j]) for j in range(0, 171) ])
print("Finalizado a Conversao")
```

### devido a demora na conversão parace necessário salvar em txt a lista.

```python
arq = open('listTransacoes.txt', 'w')
arq.write(str(ListaTransacoes))
arq.close()
```

```python
ListaTransacoes
```

# 2. Utilização do Apyori como biblioteca do apriori

# 3. Considerando dados

### Considerando que trata-se de um supermercado o periodo minimo de interesse deve ser de uma semana (7 dias) considerando que uma frequência diária de interesse 5 vezes ao dia, logo 7* 5 = 35 vezes contidos em 1715 trasações, confiança 0,02 aproximamente.

```python
mais = apriori(ListaTransacoes, min_support=0.02, min_length=1)  
mais_result = list(mais)  
```

### Lista de itens frequentes

```python
mais_result
def sortSecond(val): 
    return val[1] 
mais_result.sort(key = sortSecond, reverse = True) 
```

### Numero de itens frequentes acima de 

```python
print(len(mais_result))  
```

### Analisando as associações geradas

```python
print("==Itens Mas Frequentes===")
for item in mais_result:

    
    pair = item[0] 
    items = [x for x in pair]
    print("Item: " + items[0])

    print("Suporte: " + str(item[1]))
    print("=====================================")
```

### Agora vamos analizar para 2 produtos

### Nessas regras vemos regras triviais

```python
regra2 = apriori(ListaTransacoes, min_support=0.02,  min_lift=1.1, min_confidence=0.6 ,min_length=2, max_length=2)  
regra2_result = list(regra2)  

def sortSecond(val): 
    return val[1] 
regra2_result.sort(key = sortSecond, reverse = True)
regra2_result
```

```python
print("==Regras===")
for item in regra2_result:

    
    pair = item[0] 
    items = [x for x in pair]
    print("Item: " + items[0] + " -> " + items[1])

    print("Suporte: " + str(item[1]))
    print("Confiança: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
