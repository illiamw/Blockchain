# Agrupamento de dados

#### Ao aplicaros dois algoritmos poderemos ter uma versão geral da diferenças entre o calculo das distância e normalização realizada por cada algoritmo (Kmeans-3 e Hierarquico). 
### Objetivo
#### O objetivo desta atividade é proporcionar uma visão geral do aprendizado não-supervisionado, em especial dos métodos de agrupamento.

## Incluindo biblioteca utilizadas para os alogoritmos citados a cima 

```python
import pandas
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import dendrogram, linkage
from mpl_toolkits.mplot3d import Axes3D
```

## Reconhendo os dados
###  Neste banco de dados existem 16 palavras (em inglês), na qual cada uma delas possui 300 atribuitos que as caracterizam no espaço vetorial previamente elaborada (Pré-processamento), por menores não conhecemos os critérios, porém para os fins dessa atividade considera-se o espaço vetorial bem definido.

```python
data = pandas.read_csv('words_data.csv', header=None, index_col=0)
```

```python
data.head()
```

### Utilizando a biblioteca sklearn.cluster.Kmaens e como sugerido 3 clusters a serem definidos

```python
kmaens = KMeans(n_clusters= 3, random_state=0).fit(data.values)
```

### Apresenta-se os grupos gerados pelo Kmeans-3

```python
grupos = pandas.DataFrame()
grupos['Palavras'] = data.index.values
grupos['Grupos'] = kmaens.labels_
grupos.sort_values(by=['Grupos'])
```

#### Grupo 0 (Automobilistico, corrida: ações, objetos, consequência):
car, bus, drive, road, fire, dead, circuit.
#### Grupo 1 (Jogos esportivos):
golf, sports, racing, soccer, game, rugby.
#### Grupo 2 (Obscuridade):
Devil, Hell, dark. 

*Observação - No Grupo 0 Dead tambem poderia se enquadrar facilmente no Grupo 2 (Obscuridade) no sentido semântico.
Veremos a seguir se o agrupamento hierarquico emividência essa possivél sobrepossição de grupo na palavra "Dead".*

### Gerando os dados em dimensão 3, obteve-se os 3 vetores com o método fit_transform

```python
axes =kmaens.fit(data).transform(data)
```

```python
axes1 = []
for i in range(len(axes)):
    axes1.append(axes[i][0])
axes2 = []
for i in range(len(axes)):
    axes2.append(axes[i][1])
axes3 = []
for i in range(len(axes)):
    axes3.append(axes[i][2])
```

```python
fig = plt.figure(figsize=(40, 20))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(axes1,axes2,axes3, c=kmaens.labels_, linewidths=50 )
```

### Aplicando o algoritmo Hierarquico da bilioteca scipy.cluster.hierarchy geraremos os grupos e visualizaremos com um dandrograma

```python
hierarquico = linkage(data, 'ward')
```

```python
fig = plt.figure(figsize=(10, 10))
plt . title ( 'Dendrograma de Agrupamento Hierárquico' ) 
plt . xlabel ( 'palavras' )
plt . ylabel ( 'distância' ) 
dn = dendrogram(hierarquico)
plt. show ()
```

### Comparando os grupos gerados pelo Kmeans-3 e Hierarquico

```python
comparar = pandas.DataFrame()
```

```python
comparar["Palavras"] = data.index.values
comparar["Kmeans-3"] = kmaens.labels_
comparar["Hierarquico"] = [0,0,0,0,2,0,2,2,2,1,1,0,0,1,0,1]
comparar.set_index('Palavras')
comparar.sort_values(by =['Kmeans-3','Hierarquico'])
```

### Taxa de coincidência entre Algoritmos

```python
match = 0
for i in range(len(comparar)):
    if(comparar.iloc[i][1] == comparar.iloc[i][2]):
        match = match + 1
porc_match =(match / len(comparar))*100
porc_match
```

### Analise
#### As palavras que se destacam nesses dois métodos de Agrupamento é "Dead" e "Game", em um contexto semântico podemos associar "Dead" tanto ao Grupo 0 (automobilistico) quanto Grupo 2 (obscuridade) para kmeans-3, ja o agrupamento hierarquico a palavra "Dead" esta associado a Grupo 2 (Obscuridade), porém a palavra "Game" a Grupo 0 (Game de corrida), no agrupamento hierarquico, ao contrário do agrupamento Kmeans-3 que foi agrupado em Grupo 1 (Jogos esportivos).
### Conclusão
#### O significado semântico do grupo pode alterar conforme o método de agrupamento, podemos observar no Grupo 0 (Automobilistico) e Grupo 0 (Games de corrida), se o viés dos significado dos grupos não estiver incorporado na amostragem das palavras os conjuntos podem gerar contextos diferentes da mesma palavra, como Game (categoria jogos) ou Game (sinalizando um grupo de palavras distinguindo do as de automobilistomo de jogos de corridas).
